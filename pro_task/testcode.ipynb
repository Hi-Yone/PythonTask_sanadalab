```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from sympy.plotting import plot

def normal_distlibution(x, mean, sigma):
    y = np.exp(-(x - mean)**2 / 2*(sigma**2))
    return y

def FirstOrder_differential(x, mean, sigma):
    y = np.exp(-(x - mean)**2 / 2*(sigma**2))
    yd = sym.Derivative(y).doit()
    return yd

x, y, mean, sigma = sym.symbols("x y mean sigma")
x = np.arange(-5, 5, 0.1)
fig = plt.figure()

ax1 = plt.subplot(2,2,1)
ax1.grid()
ax1.plot(x, normal_distlibution(x, 0, 1))

ax2 = plt.subplot(2,2,2)
ax2.grid()
ax2.plot(x, FirstOrder_differential(x, 0, 1))

ax3 = plt.subplot(2,2,3)
ax3.grid()
ax3.plot(x, normal_distlibution(x, 0, 1))
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.close("all")

# =============================================================================
# Generate data
# =============================================================================
# Smpling frequency (Nyquist Freq = sampling_rate/2 [Hz]）
sampling_freq = 64 # [Hz]
# sample time [s]
t = np.arange(-1, 1, 1/sampling_freq)
N = np.size(t)
dt = 1/sampling_freq

freq = 4 # Frequency
amp = 1 # Amplitude
f1 = amp * np.sin(2*np.pi*freq*t) # Generate signal
freq = 8 # Frequency
amp = 2 # Amplitude
f2 = amp * np.sin(2*np.pi*freq*t) # Generate signal
f = f1 + f2

# =============================================================================
# # Conduct Fast Fourier transform
# =============================================================================
F = np.fft.fft(f) #
# compute absolute value of FFT
F_abs = np.abs(F)
# Adjust the amplitude to original signal wave
F_abs_amp = F_abs / N * 2 # subtruct by number of data and double the value
F_abs_amp[0] = F_abs_amp[0] / 2 #DC component must be divided by 2

shifted_F_abs_amp = np.fft.fftshift(F_abs_amp)
# generage frequency axis
freqs = np.fft.fftfreq(N, dt)
shifte_ferq = np.fft.fftshift(freqs)


# =============================================================================
# # show graph
# =============================================================================
fig = plt.figure(figsize=(12, 4))
plt.subplots_adjust(wspace=0.4, hspace=0.7)
# =============================================================================
# # plot signal
# =============================================================================
ax2 = fig.add_subplot(2,2,1)
plt.xlabel('time(sec)', fontsize=14)
plt.ylabel('amplitude', fontsize=14)
plt.plot(t, f)
plt.ylim([-3,3])
# =============================================================================
# # plot amplitude spectrum
# =============================================================================
ax2 = fig.add_subplot(2,2,2)
plt.plot(freqs, F_abs_amp,'r.-')# amplitude spectrum, before fftshift
plt.plot(shifte_ferq,shifted_F_abs_amp,'b.-')
plt.xlabel('freqency(Hz)', fontsize=14)
plt.ylabel('amplitude', fontsize=14)
plt.ylim([0,3])

# =============================================================================
#  Inverse FFT
# =============================================================================
F_ifft = np.fft.ifft(F) # inverse FFT
F_ifft_real = F_ifft.real # real
# F_ifft_real[:10]
# =============================================================================
# # Display reconstructed data
# =============================================================================
ax2 = fig.add_subplot(2,2,3)
plt.plot(t, F_ifft_real, c="g") # plot inversed data
plt.ylabel('amplitude', fontsize=14)
plt.ylim([-3,3])
plt.title('reconstructed function')
# =============================================================================
# Display difference of original data and reconstructed data
# =============================================================================
ax2 = fig.add_subplot(2,2,4)
f_diff = f - F_ifft_real
plt.plot(t, f_diff, c="r") # plot inversed data
plt.ylim([-3,3])
plt.ylabel('amplitude', fontsize=14)
plt.title('difference')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy

amp1, freq1, phase1, baseline1 = 1, 0.2, 0, 0.5                   # Amplitude, Frequency, Phase, Baseline
amp2, freq2, phase2, baseline2 = 2, 2, 270, 0                     # Amplitude, Frequency, Phase, Baseline

phase_rad1, phase_rad2 = phase1*np.pi/180, phase2*np.pi/180     # 位相
omega1, omega2 = 2*np.pi*freq1, 2*np.pi*freq2   # 角周波数
sampling_freq = 60                              # サンプリング周波数

x_time = np.arange(-5, 5.01, 1/sampling_freq)   # データポイント生成

# sin波
def sinewave(timedata, amp, freq, phase, baseline):
    phase_rad = phase*np.pi/180     # 位相
    omega = 2*np.pi*freq            # 角周波数
    y = amp * np.sin(omega * timedata - phase_rad) + baseline
    return y

def diff4(x, h):
    """
    1回微分{-f(x+2h)+8f(x+h)-8f(x-h)+f(x-2h)}/12h
    xは時系列データ,　hはデータ間の時間(second)
    """
    res = -x[4:] + 8*x[3:-1] - 8*x[1:-3] + x[:-4]
    return res/(12*h)

y1 = sinewave(x_time, amp1, freq1, phase1, baseline1)
dy = diff4(y1, x_time[1] - x_time[0])
dy_where = x_time[np.where((dy >= -0.01) & (dy<0.01))]
print(dy_where)
```

```python
import numpy as np
import matplotlib.pyplot as plt


# my function to generate histograms. THis function returns x axis, and histogram data
def my_histogram(xmin, xmax, bbin ,data):
    bins_array = np.arange(xmin-bbin,xmax+bbin,bbin)
    hist_, bins_ = np.histogram(data, bins=bins_array, range=(xmin, xmax),
                           normed=False, weights=None)
    bins_centr = bins_[0:len(bins_)-1] + bbin/2
    print("hist_ : ", len(bins_array))
    return(bins_centr, hist_, bins_)

def find_datapoint_passed_threshold(data,threshold):
    NN = len(data)
    yy = data
    yy_prev = np.append( yy[0],yy[0:len(yy)-1] )
    # get location of data which passed criteria you set.
    select_ii = np.where((yy>=threshold) & (yy_prev<threshold))
    selected_yy = np.zeros(NN)
    # to visualize which elements are selected, change value from 0 to 1
    selected_yy[select_ii] = 1
    return selected_yy, select_ii


def extract_spikewave(spk_wave , spk_t, ttime, timewin):
    ttime2 = np.round(ttime, decimals=1)
    spk_t2 = np.round(spk_t, decimals=1)
    incr = ttime[1]-ttime2[0]
    tmp = np.arange(timewin[0],timewin[1]+incr,incr, dtype=np.float)
    tt =  np.linspace(timewin[0], timewin[1],len(tmp))
    spkwave_extracted = np.zeros((len(spk_t), len(tmp)))
    for ii in range(len(spk_t2)):
        temp_ttime = ttime2- spk_t2[ii]
        norm_ttime = temp_ttime[np.where((temp_ttime>=timewin[0]) &  (temp_ttime<=timewin[1]))]
        norm_ttime = np.round(norm_ttime,decimals=1)
        
        curr_spk_wave = spk_wave[np.where((temp_ttime>=timewin[0]) &  (temp_ttime<=timewin[1]))]
        spk_wave_temp = np.zeros(len(tt))
        
        norm_ttime0 = np.round(norm_ttime[0],decimals=1)
        if (len(norm_ttime)<len(tt)) and (timewin[0] < norm_ttime[0]):
            first_point = np.where(tt == norm_ttime[0])
            spk_wave_temp[int(first_point[0]):int(first_point[0]) + len(curr_spk_wave)]= curr_spk_wave
        elif (len(norm_ttime)<len(tt)) and  (norm_ttime[len(norm_ttime)-1] < timewin[1]):
            spk_wave_temp[0:len(norm_ttime)] = curr_spk_wave
        else:
            spk_wave_temp = curr_spk_wave
        spkwave_extracted[ii]= spk_wave_temp
    return spkwave_extracted, tt


fsz = 16
# load array
loaded_array = np.load('DetectSpike/Task_spike_wave_data.npz')
spike_waveform = loaded_array['spkdata']
ttime = loaded_array['timedata']
stimON = loaded_array['stimON']
stimOFF = loaded_array['stimOFF']
threshold = 1.5 # Volt

#----------------------------
plt.close("all")
mm = 3
nn = 1
fig_ = plt.figure(figsize=(8,10))
plt.subplots_adjust(wspace=0.4, hspace=0.6)
# =============================================================================
# Plot data and stim timing
# =============================================================================
plt.subplot(mm,nn,1)
plt.plot(ttime, spike_waveform,'b-',label = 'spike waveform')
plt.plot([stimON, stimON], [-3, 3],'r--',label = 'STIM ON')
plt.plot([stimOFF, stimOFF], [-3, 3],'m--',label = 'STIM OFF')
plt.ylabel('voltage(V)', fontsize=fsz)
plt.xlabel('time from STIM ONSET (msec)', fontsize=fsz)
plt.ylim([-3,3])
plt.xlim([ttime[0],ttime[len(ttime)-1]])
plt.plot(ttime, np.ones(len(ttime))*threshold,'g--',label = 'threshold')

# =============================================================================
# find datapoint 
# =============================================================================
selected_yy, select_ii = find_datapoint_passed_threshold(spike_waveform,threshold)
# plot spike timing
spk_timing_t = ttime[select_ii]
# Extract 
spk_extracted, tt_spk_extracted = extract_spikewave(spike_waveform, spk_timing_t, ttime, [-3,4])
                                        
spk_timing_y = np.ones(len(spk_timing_t))*2
plt.plot(spk_timing_t,spk_timing_y,'rx',label = 'spike detected')
plt.xlim([ttime[0],ttime[len(ttime)-1]])
plt.legend(loc='best', fontsize = 6)

plt.grid

# =============================================================================
# Histogram
# =============================================================================
time_bin = 20
tmin = ttime[0]
tmax = ttime[len(ttime)-1]
[bb, y_hst, bins_] =  my_histogram(tmin, tmax, time_bin , spk_timing_t)
# Compute firing rate
Firing_rate = y_hst/(time_bin/1000) # 20(msec)/1000(msec) = 0.01(sec)
plt.subplot(mm,nn,2)
plt.plot([stimON, stimON], [0, max(Firing_rate)*1.1],'r--')
plt.plot([stimOFF, stimOFF], [0, max(Firing_rate)*1.1],'m--')
plt.bar(bb, Firing_rate, color='blue', width=time_bin)
plt.ylim([0,max(Firing_rate)*1.1])
plt.xlim([ttime[0],ttime[len(ttime)-1]])
plt.ylabel('Firing rate (spikes/sec) \n (spike number divided by {0:.4g}msec)'.format(time_bin), fontsize=fsz)
plt.xlabel('time from STIM ONSET (msec)', fontsize=fsz)
plt.title('PSTH (Post Stimulus Time Histogram, bin={0:.4g}'.format(time_bin) +'(msec)', fontsize=fsz)
plt.grid

# ============================================================================= 
# Display extracted spike waves
# =============================================================================
plt.subplot(mm,nn,3)
plt.plot(tt_spk_extracted,spk_extracted.T,'b-')
plt.plot([0,0],[-3,3],'r--')
plt.ylim([-3,3])
plt.plot([-3,4], np.ones(2)*threshold,'g--')

plt.title('Spike Waveform', fontsize=fsz)
plt.ylabel('Voltage(V)', fontsize=fsz)
plt.xlabel('normalized time (msec)', fontsize=fsz)
plt.grid()
plt.xlim([-2,2])
plt.show
```
