```python
import numpy as np
import matplotlib.pyplot as plt

NN = 1000
freq = .2
# generate data : yy = sine + noise
xx = 4*np.random.rand(NN)
yy = 3*np.sin(2*np.pi*freq*xx) + np.random.randn(NN)*0.5

# divide data into small group and compute standarad deviantion and mean value.
bbin = .4
minval = 0
maxval = 4
xx_binned = np.arange(minval,maxval+bbin, bbin)
meanval = np.empty(len(xx_binned))
std_val = np.empty(len(xx_binned))
for ii in range(len(xx_binned)):
    temp = yy[np.where((xx >= xx_binned[ii] - bbin/2) & (xx < xx_binned[ii] + bbin/2))]
    meanval[ii] = np.mean(temp)
    std_val[ii] = np.std(temp)


# plot original data and mean+/-std by errorbar plot
plt.close("all")
plt.figure
plt.plot(xx,yy,'c.',label='y=sine+noise')
plt.errorbar(xx_binned, meanval, yerr = std_val, capsize=5, fmt='o', markersize=10, ecolor='black', markeredgecolor = "black", color='w',label='mean+/-std')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from sympy.plotting import plot

def normal_distlibution(x, mean, sigma):
    y = np.exp(-(x - mean)**2 / 2*(sigma**2))
    return y

def FirstOrder_differential(x, mean, sigma):
    y = np.exp(-(x - mean)**2 / 2*(sigma**2))
    yd = sym.Derivative(y).doit()
    return yd

x, y, mean, sigma = sym.symbols("x y mean sigma")
x = np.arange(-5, 5, 0.1)
fig = plt.figure()

ax1 = plt.subplot(2,2,1)
ax1.grid()
ax1.plot(x, normal_distlibution(x, 0, 1))

ax2 = plt.subplot(2,2,2)
ax2.grid()
ax2.plot(x, FirstOrder_differential(x, 0, 1))

ax3 = plt.subplot(2,2,3)
ax3.grid()
ax3.plot(x, normal_distlibution(x, 0, 1))
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.close("all")

# =============================================================================
# Generate data
# =============================================================================
# Smpling frequency (Nyquist Freq = sampling_rate/2 [Hz]
sampling_freq = 64 # [Hz]
# sample time [s]
t = np.arange(-1, 1, 1/sampling_freq)
N = np.size(t)
dt = 1/sampling_freq

freq = 4 # Frequency
amp = 1 # Amplitude
f1 = amp * np.sin(2*np.pi*freq*t) # Generate signal
freq = 8 # Frequency
amp = 2 # Amplitude
f2 = amp * np.sin(2*np.pi*freq*t) # Generate signal
f = f1 + f2

# =============================================================================
# # Conduct Fast Fourier transform
# =============================================================================
F = np.fft.fft(f) #
# compute absolute value of FFT
F_abs = np.abs(F)
# Adjust the amplitude to original signal wave
F_abs_amp = F_abs / N * 2 # subtruct by number of data and double the value
F_abs_amp[0] = F_abs_amp[0] / 2 #DC component must be divided by 2

shifted_F_abs_amp = np.fft.fftshift(F_abs_amp)
# generage frequency axis
freqs = np.fft.fftfreq(N, dt)
shifte_ferq = np.fft.fftshift(freqs)


# =============================================================================
# # show graph
# =============================================================================
fig = plt.figure(figsize=(12, 4))
plt.subplots_adjust(wspace=0.4, hspace=0.7)
# =============================================================================
# # plot signal
# =============================================================================
ax2 = fig.add_subplot(2,2,1)
plt.xlabel('time(sec)', fontsize=14)
plt.ylabel('amplitude', fontsize=14)
plt.plot(t, f)
plt.ylim([-3,3])
# =============================================================================
# # plot amplitude spectrum
# =============================================================================
ax2 = fig.add_subplot(2,2,2)
# plt.plot(freqs, F_abs_amp,'.-')# amplitude spectrum, before fftshift
plt.plot(shifte_ferq,shifted_F_abs_amp,'b.-')
plt.xlabel('freqency(Hz)', fontsize=14)
plt.ylabel('amplitude', fontsize=14)
plt.ylim([0,3])

# =============================================================================
#  Inverse FFT
# =============================================================================
F_ifft = np.fft.ifft(F) # inverse FFT
F_ifft_real = F_ifft.real # real
# F_ifft_real[:10]
# =============================================================================
# # Display reconstructed data
# =============================================================================
ax2 = fig.add_subplot(2,2,3)
plt.plot(t, F_ifft_real, c="g") # plot inversed data
plt.ylabel('amplitude', fontsize=14)
plt.ylim([-3,3])
plt.title('reconstructed function')
# =============================================================================
# Display difference of original data and reconstructed data
# =============================================================================
ax2 = fig.add_subplot(2,2,4)
f_diff = f - F_ifft_real
plt.plot(t, f_diff, c="r") # plot inversed data
plt.ylim([-3,3])
plt.ylabel('amplitude', fontsize=14)
plt.title('difference')
```
